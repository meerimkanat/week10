Abroad Study Cost Predictor.
This dataset provides information about the yearly fees for various courses offered by colleges and universities across different countries. I used this data from Kaggle https://www.kaggle.com/datasets/shivampawale/abroad-study-cost-predictor.The goal is to uncover insights about the cost of studying abroad, particularly focusing on Germany.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file with specified column names
file_path = 'abroad.csv'  # Replace with the actual file path if necessary
column_names = ['COUNTRY', 'COURSE TYPE', 'COURSE (SPECIALIZATION)', 'FEES']
data = pd.read_csv(file_path, names=column_names, skiprows=1)  # Skip the first row of data

# Convert the 'FEES' column to numeric type
data['FEES'] = pd.to_numeric(data['FEES'], errors='coerce')  # Convert to numbers, non-numeric replaced with NaN

# Filter data for Germany
germany_data = data[data['COUNTRY'] == 'GERMANY']

# Check the filtered data
print("Data for Germany:\n", germany_data.head())

# Average course fees in Germany
avg_fees_by_course = germany_data.groupby('COURSE (SPECIALIZATION)')['FEES'].mean()

# Select the top 10 courses with the highest average fees
top_10_courses = avg_fees_by_course.nlargest(10)

# Check for any NaN values
print("Top 10 courses with the highest average fees:\n", top_10_courses)

# Plot a bar chart
plt.figure(figsize=(10, 6))
top_10_courses.plot(kind='bar', color='skyblue')
plt.title('Top 10 Courses with Highest Average Tuition Fees in Germany')
plt.xlabel('Course Specialization')
plt.ylabel('Average Fees')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()

# Plot a pie chart
plt.figure(figsize=(8, 8))  # Set figure size

# Create the pie chart
plt.pie(
    top_10_courses, 
    labels=top_10_courses.index,  # Course names
    autopct='%1.1f%%',  # Show percentages
    startangle=90,  # Starting angle
    colors=plt.cm.Paired.colors,  # Use paired color scheme
    wedgeprops={'edgecolor': 'black'}  # Add black edge to segments
)

# Add a title
plt.title('Top 10 Courses with Highest Average Tuition Fees in Germany')

# Show the plot
plt.show()
```
